```python

import dash
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import plotly.express as px
from dash.dependencies import Input, Output
```

```python
df1 = pd.read_csv('e:/archive/country_wise_latest.csv')
df2 = pd.read_csv('e:/archive/covid_19_clean_complete.csv')
df3 = pd.read_csv('e:/archive/worldometer_data.csv')
df4 = pd.read_csv('e:/archive/usa_county_wise.csv')
df5 = pd.read_csv('e:/archive/full_grouped.csv')
df6 = pd.read_csv('e:/archive/day_wise.csv')
```

```python
df2['Date'] = pd.to_datetime(df2['Date'])

app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("COVID-19 Data Dashboard"),
    dcc.Dropdown(
        id='country-dropdown',
        options=[{'label': country, 'value': country} for country in df2['Country/Region'].unique()],
        value='World',
        clearable=False
    ),
    dcc.Graph(id='covid-trends'),
    dcc.Graph(id='worldometer-stats')
])

@app.callback(
    [Output('covid-trends', 'figure'),
     Output('worldometer-stats', 'figure')],
    [Input('country-dropdown', 'value')]
)
def update_graphs(selected_country):
   
    if selected_country is None or selected_country == 'World':
        df_selected = df2
    else:
        df_selected = df2[df2['Country/Region'] == selected_country]

    if df_selected.empty:
        fig_trends = px.line(title=f"No data available for {selected_country}")
    else:
     
        fig_trends = px.line(
            df_selected, x='Date', y=['Confirmed', 'Deaths', 'Recovered'],
            title=f"COVID-19 Trends in {selected_country}"
        )
    fig_worldometer = px.bar(
        df3, x='Country/Region', y=['TotalCases', 'TotalDeaths', 'TotalRecovered'],
        title="Worldometer Stats"
    )

    return fig_trends, fig_worldometer

if __name__ == '__main__':
    app.run_server(debug=True, port=8052)  # Use a different port number

```
